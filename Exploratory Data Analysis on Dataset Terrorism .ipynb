## Top 10 countrieswith most terrorist attack

```python
print(df['country_txt'].value_counts().head(10))
```

```python
countries = df['country_txt'].value_counts()[:10].index
values = df['country_txt'].value_counts()[:10].values
plt.figure(figsize=(15,10))
s = pd.Series(values,index = countries).plot(kind = 'barh',color = 'red')
plt.title('Countries with most terrorist attack',fontsize = 25)
```

#### 1.Most Affected country:iraq
#### 2.Among top 10 countries 5 are asian

## Weapons used : Worldwide VS India

```python
weapons_worldwide = df['weaptype1_txt'].value_counts()[:5].index
values_weapons_worldwide = df['weaptype1_txt'].value_counts()[:5].values

plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.bar(weapons_worldwide, values_weapons_worldwide, color = 'red')
plt.title('Weapons mostly used worldwide',fontsize = 25)
plt.show()

weapons_india = df[df['country_txt']=='India']['weaptype1_txt'].value_counts()[:5].index
values_weapons_india = df[df['country_txt']=='India']['weaptype1_txt'].value_counts()[:5].values

plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.bar(weapons_india, values_weapons_india, color = '#03a9f4')
plt.title('Weapons mostly used in India',fontsize = 25)
plt.show()
```

#### Most weapon used : Worldwide VS India : explosives


## Types of target : Worldwide VS India

```python
targets_worldwide = df['targtype1_txt'].value_counts()[:5].index
values_targets_worldwide = df['targtype1_txt'].value_counts()[:5].values

plt.figure(figsize=(19,10))
plt.subplot(1,2,1)
plt.bar(targets_worldwide, values_targets_worldwide, color = 'red')
plt.title('targets mostly used worldwide',fontsize = 25)
plt.show()

targets_india = df[df['country_txt']=='India']['targtype1_txt'].value_counts()[:5].index
values_targets_india = df[df['country_txt']=='India']['targtype1_txt'].value_counts()[:5].values

plt.figure(figsize=(19,10))
plt.subplot(1,2,1)
plt.bar(targets_india, values_targets_india, color = '#03a9f4')
plt.title('targets mostly used in India',fontsize = 25)
plt.show()
```

#### Most targeted : Worldwide VS India : private citizens and property


## Top 3 terrorist group involved : Worldwide VS India

```python
groups_worldwide = df['gname'].value_counts()[:5].index
values_groups_worldwide = df['gname'].value_counts()[:5].values

plt.figure(figsize=(19,10))
plt.subplot(1,2,1)
plt.bar(groups_worldwide, values_groups_worldwide, color = 'red')
plt.title('targets mostly used worldwide',fontsize = 20)
plt.show()

groups_india = df[df['country_txt']=='India']['gname'].value_counts()[:5].index
values_groups_india = df[df['country_txt']=='India']['gname'].value_counts()[:5].values

plt.figure(figsize=(19,10))
plt.subplot(1,2,1)
plt.bar(groups_india, values_groups_india, color = '#03a9f4')
plt.title('groups mostly used in India',fontsize = 20)
plt.show()
```
